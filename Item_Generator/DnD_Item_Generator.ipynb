```python
import os
from PIL import Image
import torch
from diffusers import StableDiffusionPipeline
import clip
```

imports

```python
# Root folder of your dataset
DATASET_ROOT = r"C:\Users\DSU\Desktop\Tanner_Allen_Text_Image\dataset"

# Folders for images and text
OUTPUT_DIR = os.path.join(DATASET_ROOT, "generated_items")
os.makedirs(OUTPUT_DIR, exist_ok=True)

device = "cuda" if torch.cuda.is_available() else "cpu"
print("Using device:", device)
```

set data folder path, set output file path, set evice to cpu

```python
# Load Stable Diffusion
pretrained_model = "runwayml/stable-diffusion-v1-5"
pipe = StableDiffusionPipeline.from_pretrained(pretrained_model)
pipe = pipe.to(device)

# Load CLIP
clip_model, clip_preprocess = clip.load("ViT-B/32", device=device)
```

load models for diffusion and clip

```python
# Loop through dataset
for category in os.listdir(DATASET_ROOT):
    cat_path = os.path.join(DATASET_ROOT, category)
    if not os.path.isdir(cat_path):
        continue
    
    images_folder = os.path.join(cat_path, "images")
    captions_folder = os.path.join(cat_path, "captions")
    
    if not os.path.exists(images_folder) or not os.path.exists(captions_folder):
        continue
    
    for img_file in os.listdir(images_folder):
        if not img_file.lower().endswith((".png", ".jpg", ".jpeg")):
            continue
        
        # Corresponding text file
        txt_file = os.path.splitext(img_file)[0] + ".txt"
        txt_path = os.path.join(captions_folder, txt_file)
        
        if not os.path.exists(txt_path):
            continue
        
        # Read text description
        with open(txt_path, "r", encoding="utf-8") as f:
            prompt = f"a {f.read().strip()} magical D&D item, detailed fantasy illustration"
        
        # Open image for reference
        ref_image_path = os.path.join(images_folder, img_file)
        ref_image = Image.open(ref_image_path).convert("RGB")
        
        # Generate image using Stable Diffusion, strength means how much image affects generated image 0-1
        generated_image = pipe(prompt, init_image=ref_image, strength=0.7).images[0]
        
        # Save generated image
        output_path = os.path.join(OUTPUT_DIR, f"{category}_{os.path.splitext(img_file)[0]}_gen.png")
        generated_image.save(output_path)
        print("Generated:", output_path)
```

use clip and a rag like format to pass clip information, the use diffusion to create image
